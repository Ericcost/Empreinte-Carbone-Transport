```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('HBFA_V4.1 2000_2050.csv')

df
```

```python
# Assuming 'polluant' is the name of the column in your DataFrame
polluant_dummies = pd.get_dummies(df['Polluant'], prefix='polluant')

# Merge the dummy columns with the original DataFrame
df_with_polluants = pd.concat([df, polluant_dummies], axis=1)

# Print the first few rows to check
df_with_polluants.head()
```

```python
# Step 1: Create a new column with the "Polluant" + "Unite" combination as the new column name
df['Polluant_Unite'] = df['Polluant'] + "_" + df['Unite']

# Step 2: Pivot the table to reshape it
df_pivot = df.pivot_table(index=['Pays', 'Annee', 'Categorie de vehicules'], 
                          columns='Polluant_Unite', 
                          values='emission facteur', 
                          aggfunc='first')  # Use first in case of duplicates

# Step 3: Reset the index to flatten the DataFrame
df_pivot_reset = df_pivot.reset_index()

# Print the first few rows to check
df_pivot_reset
```
